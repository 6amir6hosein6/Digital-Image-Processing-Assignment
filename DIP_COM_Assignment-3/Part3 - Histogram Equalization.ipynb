### 3. Histogram Equalization

#### a. Histogram Equalization Techniques

**Histogram Equalization (HE)**: HE enhances overall image contrast by adjusting the global pixel intensity distribution. It spreads out frequent pixel values, resulting in improved contrast.

<br>

**Adaptive Histogram Equalization (AHE)**: AHE operates locally on small regions (tiles) within the image. It adapts the contrast to specific areas, making it effective for uneven lighting or localized variations . 

#### b. Load and Display Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

```python
img = cv2.imread('moon.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(histogram)
plt.show()
```

#### c. Simple Histogram Equalization

```python
img_eq = exposure.equalize_hist(img)

plt.imshow(img_eq, cmap='gray')
plt.show()
```

#### d. Adaptive Histogram Equalization

```python
clahe = cv2.createCLAHE(clipLimit=30, tileGridSize=(7, 7))

output_img = clahe.apply(img)

plt.imshow(output_img, cmap='gray')
plt.show()
```
