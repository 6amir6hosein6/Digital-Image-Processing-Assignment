```python
filter1 = np.array([1, 0, -1])

result = convolve2d(image, filter1.reshape((1, filter1.shape[0])), mode='same')

plt.imshow(result, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
filter1 = np.array([1,-1])

result = convolve2d(image, filter1.reshape((filter1.shape[0], 1)), mode='same')

plt.imshow(result, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
filter1 = np.array([[-1,-1, -1],[-1,8, -1],[-1,-1, -1]])

result = convolve2d(image, filter1, mode='same')

plt.imshow(result, cmap='gray', vmin=0, vmax=255)
plt.show()
```

**In the first filter**, we can see that applying the filter makes some edges visible. However, it primarily highlights the edges that change from black to white. As observed in the result, this filterâ€™s effect on converting white to black and enhancing horizontal edges is not very pronounced.

**The second filter** did not significantly affect the original image; it merely returned the individual image.

<br>

**The third filter** has an effect similar to the first filter but introduces some boldness to the edged image.

<br>

**The fourth and last filter**, however, performs much better than the others. It makes all edgesâ€”black to white, white to black, and both horizontal and verticalâ€”much more apparent. This filter is actually a Laplacian filter, which, when applied to an image, effectively detects edges.
