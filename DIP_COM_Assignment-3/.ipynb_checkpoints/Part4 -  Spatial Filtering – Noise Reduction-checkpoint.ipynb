#### b. Load and Display Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
from scipy.ndimage import median_filter
from scipy.ndimage import uniform_filter
from scipy.ndimage import gaussian_filter
```

```python
co = cv2.imread('Images/city_orig.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(co, cmap='gray', vmin=0, vmax=255)
plt.show()

cn = cv2.imread('Images/city_noise.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cn, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
x= int(co.shape[0]/2)
y= int(co.shape[1]/2)

co2 = co[0: x, 0: y]
co1 = co[0: x, y: y * 2]
co4 = co[x: x * 2 , 0:y]
co3 = co[x: x * 2, y: y * 2]


cn2 = cn[0: x, 0: y]
cn1 = cn[0: x, y: y * 2]
cn4 = cn[x: x * 2 , 0:y]
cn3 = cn[x: x * 2, y: y * 2]


plt.figure(figsize=(15, 15))  # Set the desired width and height

plt.subplot(4,2,1)
plt.imshow(co2, cmap='gray')
plt.subplot(4,2,2)
plt.imshow(co1, cmap='gray')
plt.subplot(4,2,3)
plt.imshow(co4, cmap='gray')
plt.subplot(4,2,4)
plt.imshow(co3, cmap='gray')


plt.subplot(4,2,5)
plt.imshow(cn2, cmap='gray')
plt.subplot(4,2,6)
plt.imshow(cn1, cmap='gray')
plt.subplot(4,2,7)
plt.imshow(cn4, cmap='gray')
plt.subplot(4,2,8)
plt.imshow(cn3, cmap='gray')
```

#### a. Computing SNR

```python
def SNR(img1, img2):
    total_sum = np.sum(np.power(img1,2))
    noise_sum = np.sum(np.power((img1 - img2),2)) + 0.00000001
    
    return 10 * np.log10(total_sum / noise_sum + 0.0000000001)
```

```python
print(SNR(co1, cn1))

print(SNR(co2, cn2))

print(SNR(co3, cn3))

print(SNR(co4, cn4))
```

##### Median Filter :

```python
cn2_med_filter = median_filter(cn2, size=(3, 3))
print(SNR(co2, cn2), "====>" ,SNR(co2, cn2_med_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn2_med_filter, cmap='gray')
```

```python
cn3_med_filter = median_filter(cn3, size=(3, 3))
print(SNR(co3, cn3), "====>" ,SNR(co3, cn3_med_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn3_med_filter, cmap='gray')
```

```python
cn4_med_filter = median_filter(cn4, size=(3, 3))
print(SNR(co4, cn4), "====>" ,SNR(co4, cn4_med_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn4_med_filter, cmap='gray')
```

##### Mean Filter :

```python
cn2_mean_filter = uniform_filter(cn2, size=(3, 3))
print(SNR(co2, cn2), "====>" ,SNR(co2, cn2_mean_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn2_mean_filter, cmap='gray')
```

```python
cn3_mean_filter = uniform_filter(cn3, size=(8, 8))
print(SNR(co3, cn3), "====>" ,SNR(co3, cn3_mean_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn3_mean_filter, cmap='gray')
```

```python
cn4_mean_filter = uniform_filter(cn4, size=(8, 8))
print(SNR(co4, cn4), "====>" ,SNR(co4, cn4_mean_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn4_mean_filter, cmap='gray')
```

##### Gussian Filter :

```python
sigma = 1.0
cn2_gussian_filter = gaussian_filter(cn2, sigma=sigma)
print(SNR(co2, cn2), "====>" ,SNR(co2, cn2_gussian_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn2_gussian_filter, cmap='gray')
```

```python
sigma = 3.0
cn3_gussian_filter = gaussian_filter(cn3, sigma=sigma)
print(SNR(co3, cn3), "====>" ,SNR(co3, cn3_gussian_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn3_gussian_filter, cmap='gray')
```

```python
sigma = 3.0
cn4_gussian_filter = gaussian_filter(cn4, sigma=sigma)
print(SNR(co4, cn4), "====>" ,SNR(co4, cn4_gussian_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn4_gussian_filter, cmap='gray')
```

```python
best_result = np.zeros_like(co)

best_result[0: x, 0: y] = cn2_med_filter
best_result[0: x, y: y * 2] = co[0: x, y: y * 2]
best_result[x: x * 2 , 0:y] = cn4_gussian_filter
best_result[x: x * 2, y: y * 2] = cn3_mean_filter
```
