### 3. Histogram Equalization

#### a. Histogram Equalization Techniques

#### a. Load and Display Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

```python
img = cv2.imread('Images/retina.bmp', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(histogram)
plt.show()
```

#### b. Logarithmic Transformation

```python
image_log = np.log2(img + 0.00001)
image_log_norm = cv2.normalize(image_log, None, 0, 255, cv2.NORM_MINMAX)
image_log_8bit = np.uint8(image_log_norm)
histogram = cv2.calcHist([image_log_8bit], [0], None, [256], [0, 256])


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(image_log_8bit, cmap='gray')
```

```python
image_log = np.emath.logn(10, img + 0.00001)
image_log_norm = cv2.normalize(image_log, None, 0, 255, cv2.NORM_MINMAX)
image_log_8bit = np.uint8(image_log_norm)
histogram = cv2.calcHist([image_log_8bit], [0], None, [256], [0, 256])


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(image_log_8bit, cmap='gray')
```

```python
image_log = np.log1p(img + 0.0001)
image_log_norm = cv2.normalize(image_log, None, 0, 255, cv2.NORM_MINMAX)
image_log_8bit = np.uint8(image_log_norm)
histogram = cv2.calcHist([image_log_8bit], [0], None, [256], [0, 256])


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(image_log_8bit, cmap='gray', vmin=0, vmax=255)
```

##### As evident from the resulting image and the histogram plot of the logarithmic transformation, the image becomes significantly brighter due to the logarithmic function. This transformation expands the dark pixels associated with low intensities while simultaneously reducing the distance between bright intensity values. As a consequence, the image may appear overly brightened.

#### c. Power-Law Transformation:

```python
gamma = 3
adjusted = np.power(img / 255.0, 1.0 / gamma) * 255
adjusted = np.clip(adjusted, 0, 255).astype(np.uint8)

histogram = cv2.calcHist([adjusted], [0], None, [256], [0, 256])


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(adjusted, cmap='gray', vmin=0, vmax=255)
```

```python
gamma = 1
adjusted = np.power(img / 255.0, 1.0 / gamma) * 255
adjusted = np.clip(adjusted, 0, 255).astype(np.uint8)

histogram = cv2.calcHist([adjusted], [0], None, [256], [0, 256])


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(adjusted, cmap='gray', vmin=0, vmax=255)
```

```python
gamma = 0.2
adjusted = np.power(img / 255.0, 1.0 / gamma) * 255
adjusted = np.clip(adjusted, 0, 255).astype(np.uint8)

histogram = cv2.calcHist([adjusted], [0], None, [256], [0, 256])


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(adjusted, cmap='gray', vmin=0, vmax=255)
```

```python
gamma = 0.06
adjusted = np.power(img / 255.0, 1.0 / gamma) * 255
adjusted = np.clip(adjusted, 0, 255).astype(np.uint8)

histogram = cv2.calcHist([adjusted], [0], None, [256], [0, 256])

adjusted = np.emath.logn(10, adjusted + 0.00001)
adjusted = cv2.normalize(adjusted, None, 0, 255, cv2.NORM_MINMAX)
image_log_8bit = np.uint8(adjusted)


plt.subplot(1,2,1)
plt.plot(histogram)
plt.subplot(1,2,2)
plt.imshow(adjusted, cmap='gray', vmin=0, vmax=255)
```

##### Contrary to the logarithmic transformation, the Power Law transformation expands the whiter areas. As a result, the white parts of the image can be distinguished more easily. You can observe this in the last attempt of the Power Law transformation, where the white areas contain noise and blind spots

#### d. Detection

##### An idea that comes to my mind is to first threshold the image and then count the neighborhoodness of white pixels in the image. Since the noise points do not have any neighborhood, they are not going to be counted as a blind spot

##### After counting neighborhood of each pixel take the point with largest neighborhood. That point and all of it's neighbor is the blind spot

```python
_, thresh = cv2.threshold(adjusted, 127, 255, cv2.THRESH_BINARY)
thresh = cv2.GaussianBlur(thresh, (5, 5), 0)

plane = np.zeros_like(adjusted)
result = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

areas = [cv2.contourArea(cnt) for cnt in contours]

largest_contour_index = areas.index(max(areas))
largest_contour = contours[largest_contour_index]

cv2.drawContours(result, [largest_contour], -1, (0, 255, 0), 2)
cv2.drawContours(adjusted, [largest_contour], -1, (0, 255, 0), 2)
cv2.drawContours(plane, [largest_contour], -1, (0, 255, 0), 2)

plt.subplot(1,2,1)
plt.imshow(result, cmap='gray', vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.imshow(plane, cmap='gray', vmin=0, vmax=255)
```
