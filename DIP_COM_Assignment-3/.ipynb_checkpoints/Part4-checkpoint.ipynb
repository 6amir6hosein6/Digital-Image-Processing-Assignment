```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
from scipy.ndimage import median_filter
from scipy.ndimage import uniform_filter
from scipy.ndimage import gaussian_filter
```

```python
co = cv2.imread('Helix-Nebula.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(co, cmap='gray', vmin=0, vmax=255)
plt.show()

cn = cv2.imread('noisy_salt_pepper.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cn, cmap='gray', vmin=0, vmax=255)
plt.show()
```

#### a. Computing SNR

```python
def SNR(img1, img2):
    total_sum = np.sum(np.power(img1,2))
    noise_sum = np.sum(np.power((img1 - img2),2)) + 0.00000001
    
    return 10 * np.log10(total_sum / noise_sum + 0.0000000001)
```

```python
print(SNR(cn,co))
```

##### Median Filter :

```python
cn2_med_filter = median_filter(cn, size=(3, 3))
print(SNR(co, cn), "====>" ,SNR(co, cn2_med_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn2_med_filter, cmap='gray')
```

##### Mean Filter :

```python
cn2_mean_filter = uniform_filter(cn, size=(3, 3))
print(SNR(co, cn), "====>" ,SNR(co, cn2_mean_filter))

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1)
plt.imshow(cn2_mean_filter, cmap='gray')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری تصویر خاکستری
image = cv2.imread('Phobos-Orbiter-Mars.jpg', cv2.IMREAD_GRAYSCALE)

# تابع نویز گاوسی با شدت بیشتر
def add_gaussian_noise(image, mean=0, std=50):  # std رو بیشتر کردیم
    gauss = np.random.normal(mean, std, image.shape).astype('float32')
    noisy = image.astype('float32') + gauss
    return np.clip(noisy, 0, 255).astype('uint8')

# تابع نویز نمک و فلفل با شدت بیشتر
def add_salt_pepper_noise(image, salt_prob=0.03, pepper_prob=0.03):  # بیشتر کردیم
    noisy = np.copy(image)
    total_pixels = image.size
    num_salt = int(salt_prob * total_pixels)
    num_pepper = int(pepper_prob * total_pixels)

    # Salt
    coords = [np.random.randint(0, i, num_salt) for i in image.shape]
    noisy[tuple(coords)] = 255

    # Pepper
    coords = [np.random.randint(0, i, num_pepper) for i in image.shape]
    noisy[tuple(coords)] = 0

    return noisy

# نویز ترکیبی
def add_both_noises(image):
    noisy = add_gaussian_noise(image)
    noisy = add_salt_pepper_noise(noisy)
    return noisy

# ساختن تصاویر
gaussian_noisy = add_gaussian_noise(image)
salt_pepper_noisy = add_salt_pepper_noise(image)
both_noisy = add_both_noises(image)

# ذخیره تصاویر
cv2.imwrite('noisy_gaussian.jpg', gaussian_noisy)
cv2.imwrite('noisy_salt_pepper.jpg', salt_pepper_noisy)
cv2.imwrite('noisy_both.jpg', both_noisy)

# نمایش
titles = ['Original', 'Gaussian Noise', 'Salt & Pepper Noise', 'Both Noises']
images = [image, gaussian_noisy, salt_pepper_noisy, both_noisy]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
