## Part 2 - Translate Horizontally an image via shifting its phase in frequency domain

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Reading The Library Image 

```python
image = cv2.imread(r"Library.jpg")
 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imgplot = plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Discrete Fourier Transformation (DFT)

```python
fourier = cv2.dft(np.float32(gray), flags=cv2.DFT_COMPLEX_OUTPUT)
```

### Shifting and Magnitude Calculation

```python
fourier_shift = np.fft.fftshift(fourier)
```

```python
magnitude = 20*np.log(cv2.magnitude(fourier[:,:,0],fourier[:,:,1]))
imgplot = plt.imshow(magnitude, cmap='gray')
plt.show()
```

```python
shifted_magnitude = 20*np.log(cv2.magnitude(fourier_shift[:,:,0],fourier_shift[:,:,1]))

imgplot = plt.imshow(shifted_magnitude, cmap='gray')
plt.show()
```

### The Highpass Filter

```python
filter_image = np.ones((500, 800, 2),np.uint8)
x, y = np.meshgrid(np.arange(800), np.arange(500))
distance_from_center = np.sqrt((x - 400)**2 + (y - 250)**2)


filter_image[distance_from_center <= 50] = [0, 0]

plt.imshow(filter_image[:,:,1], cmap='gray')
plt.show()

```

### Filtering

```python
filtered_image =  fourier_shift * filter_image

fmagnitude = 20*np.log(cv2.magnitude(filtered_image[:,:,0],filtered_image[:,:,1])+ 10)

plt.imshow(fmagnitude, cmap='gray')
plt.show()
```

### Decentralizing

```python
Decentralized_image = np.fft.ifftshift(filtered_image)


fDecentralized_image = 20*np.log(cv2.magnitude(Decentralized_image[:,:,0]+10,Decentralized_image[:,:,1]+10))
imgplot = plt.imshow(fDecentralized_image, cmap='gray')
plt.show()
```

### Reverse Fourier Transformation of Filterd image

```python
reversed_back = cv2.idft(Decentralized_image)

reversed_back = cv2.magnitude(reversed_back[:,:,0],reversed_back[:,:,1])

plt.imshow(reversed_back, cmap='gray')
plt.show()
```

### The Lowpass Filter

```python
filter_image = np.zeros((500, 800, 2),np.uint8)
x, y = np.meshgrid(np.arange(800), np.arange(500))
distance_from_center = np.sqrt((x - 400)**2 + (y - 250)**2)


filter_image[distance_from_center <= 50] = [1, 1]

plt.imshow(filter_image[:,:,1], cmap='gray')
plt.show()

```

```python
filtered_image =  fourier_shift * filter_image

fmagnitude = 20*np.log(cv2.magnitude(filtered_image[:,:,0],filtered_image[:,:,1])+ 10)

plt.imshow(fmagnitude, cmap='gray')
plt.show()
```

```python
Decentralized_image = np.fft.ifftshift(filtered_image)


fDecentralized_image = 20*np.log(cv2.magnitude(Decentralized_image[:,:,0]+10,Decentralized_image[:,:,1]+10))
imgplot = plt.imshow(fDecentralized_image, cmap='gray')
plt.show()
```

```python
reversed_back = cv2.idft(Decentralized_image)

reversed_back = cv2.magnitude(reversed_back[:,:,0],reversed_back[:,:,1])

plt.imshow(reversed_back, cmap='gray')
plt.show()
```

##  - Question 3:
###      What is the interpretation of the above result image ?

## + Answer:
####      The lowpass filter blocks the high-frequency parts in the image and lets the low frequencies be passed in the image. The high-frequency parts in the image are the edges of objects in the image which are due to a fast change in light intensity. Preventing the high frequencies from being passed causes the edges and high frequency to be smoother and as a result the image illustrates blurring.
