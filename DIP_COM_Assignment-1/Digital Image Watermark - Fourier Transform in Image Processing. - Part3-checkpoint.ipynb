## Part 3- Digital Image Watermark

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Read Images

```python
watermark = cv2.imread(r"KNToosi.jpg", 0)
host = cv2.imread(r"WalterBishop.jpg", 0)

plt.imshow(watermark, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(host, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Discrete Fourier Transformation (DFT)

```python
watermark = cv2.resize(watermark, (host.shape[1], host.shape[0]))


watermark_fourier = cv2.dft(np.float32(watermark), flags=cv2.DFT_COMPLEX_OUTPUT)
host_fourier = cv2.dft(np.float32(host), flags=cv2.DFT_COMPLEX_OUTPUT)
```

```python
watermark_fourier_shift = np.fft.fftshift(watermark_fourier)
host_fourier_shift = np.fft.fftshift(host_fourier)
```

```python
magnitude = 20*np.log(cv2.magnitude(watermark_fourier_shift[:,:,0],watermark_fourier_shift[:,:,1]))
imgplot = plt.imshow(magnitude, cmap='gray')
plt.show()


magnitude = 20*np.log(cv2.magnitude(host_fourier_shift[:,:,0],host_fourier_shift[:,:,1]))
imgplot = plt.imshow(magnitude, cmap='gray')
plt.show()
```

```python
weakening_coefficient = 0.02
```

```python
combined = host_fourier_shift * (1 - weakening_coefficient) + watermark_fourier_shift * weakening_coefficient

magnitude = 20*np.log(cv2.magnitude(combined[:,:,0],combined[:,:,1]))
imgplot = plt.imshow(magnitude, cmap='gray')
plt.show()
```

```python
deshift_image = np.fft.ifftshift(combined)

magnitude = 20*np.log(cv2.magnitude(deshift_image[:,:,0],deshift_image[:,:,1]))
imgplot = plt.imshow(magnitude, cmap='gray')
plt.show()
```

```python
idft_combined = cv2.idft(deshift_image)
idft_combined = cv2.magnitude(idft_combined[:,:,0],idft_combined[:,:,1])

plt.imshow(idft_combined, cmap='gray')
plt.show()
```

```python
dft_watermarked = cv2.dft(np.float32(idft_combined), flags=cv2.DFT_COMPLEX_OUTPUT)

center_x, center_y = dft_watermarked.shape[1] // 2, dft_watermarked.shape[0] // 2
x_size = 900  
y_size = 707  

dft_watermark = dft_watermarked[center_y-y_size:center_y+y_size, center_x-x_size:center_x+x_size]

idft_watermark = cv2.idft(dft_watermark)
extracted_watermark = cv2.magnitude(idft_watermark[:,:,0], idft_watermark[:,:,1])

plt.imshow(extracted_watermark, cmap='gray')
plt.show()
```
