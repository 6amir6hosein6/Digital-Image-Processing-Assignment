### Part I. Image Obliteration

#### I. Blurred Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d
```

```python
image = cv2.imread('lena_color.tif')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
def motion_blur_filter(length, angle):
    size = length
    motion_filter = np.zeros((size, size))
    
    angle = np.deg2rad(angle)
    
    center = int(np.floor(size / 2))
    
    for i in range(length):
        motion_filter[i, int(center + i * np.tan(angle))] = 1

    
    motion_filter /= motion_filter.sum()
    
    return motion_filter
```

```python
def scale_images(image):
    normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image)) * 255
    return normalized_image.astype(np.uint8)
```

### Part II. Image Restoration

#### I. Noiseless Blurred Image Inverse Filtering

```python
def inverseFiltering(blurred_image, motion_filter):
    # blurred blue channel (bb)
    # blurred green channel (bg)
    # blurred red channel (br)

    bb = blurred_image[:, :, 0]
    bg = blurred_image[:, :, 1]
    br = blurred_image[:, :, 2]

    # blurred blue channel fourier (BBF)
    # blurred green channel fourier (BGF)
    # blurred red channel fourier (BRF)

    BBF = np.fft.fft2(bb)
    BGF = np.fft.fft2(bg)
    BRF = np.fft.fft2(br)

    # motion filter (mf) zero padding to become image size

    mf = np.zeros(bb.shape) 
    mf[:motion_filter.shape[0],:motion_filter.shape[1]] = np.copy(motion_filter)

    MF = np.fft.fft2(mf)


    # F_hat in blue channel fourier (FBF)
    # F_hat in green channel fourier (FGF)
    # F_hat in red channel fourier (FRF)

    FBF = BBF / MF
    FGF = BGF / MF
    FRF = BRF / MF

    fb = np.abs(np.fft.ifft2(FBF))
    fg = np.abs(np.fft.ifft2(FGF))
    fr = np.abs(np.fft.ifft2(FRF))

    # fb /= fb.sum()

    inversed_filter_result = np.zeros_like(blurred_image)

    inversed_filter_result[:, :, 0] = fb
    inversed_filter_result[:, :, 1] = fg
    inversed_filter_result[:, :, 2] = fr

    plt.imshow(inversed_filter_result)
    plt.show()
```
