### Part I. Image Obliteration

#### a. Load and Display Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

```python
img = cv2.imread('Images/Empty_Room.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
def show_hist(image):
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.plot(histogram)
    plt.title('Intensity Histogram')
    plt.xlabel('Intensity Value')
    plt.ylabel('Frequency')
    plt.show()
```

```python
show_hist(img)
```

#### b. Dark Area Enhancement

```python
image_log = np.log1p(img + 0.00001)
image_log_norm = cv2.normalize(image_log, None, 0, 255, cv2.NORM_MINMAX)
image_log_8bit = np.uint8(image_log_norm)
show_hist(image_log_8bit)
plt.imshow(image_log_8bit, cmap='gray', vmin=0, vmax=255)
plt.show()
```

#### c. Bright Area Enhancement:

```python
gamma = 0.03
adjusted = np.power(img / 255.0, 1.0 / gamma) * 255
adjusted = np.clip(adjusted, 0, 255).astype(np.uint8)

show_hist(adjusted)
plt.imshow(adjusted, cmap='gray', vmin=0, vmax=255)
plt.show()
```
